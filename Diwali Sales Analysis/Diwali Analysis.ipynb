```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Loading the Data 

```python
df = pd.read_csv(r"C:\Users\abdul\Downloads\PROJECTS\EDA PROJECTS\Diwali Sales Analysis\Diwali Sales Data.csv", encoding='ISO-8859-1')
df.head()
```

### Exploratry Data Analysis

```python
df.head()
```

```python
df["Gender"]=df["Gender"].replace(["M","F"],["Male","Female"])
df.Marital_Status=df.Marital_Status.replace([1,0],["Unmarried","Married"])
```

#### Q Most Customer belong to which Gender (By Orders & Total Amount) ?  

```python
plt.figure(figsize=(11,5))
## 1st plot
plt.subplot(2,2,1)
cp=sns.countplot(y="Gender",data=df,palette="cool")
for bars in cp.containers:
    cp.bar_label(bars)
plt.grid(axis="y",alpha=0.5)
plt.xlabel("")
cp.set_ylabel("  Total Count",fontsize=15)
## 2nd Plot
plt.subplot(2,2,2)
x=df.groupby("Gender")["Orders"].sum()
plt.pie(x=x,labels=["Female","Male"],autopct="%1.1f%%",radius=1.2,colors=["g","r"])
x=df.groupby("Gender")["Orders"].sum()
plt.pie([1],colors="w",radius=0.5)
## 3nd plot
plt.subplot(2,2,3)
data1=df[["Gender","Amount"]].groupby(by="Gender",as_index=False).sum()
bp=sns.barplot(y="Gender",x="Amount",data=data1,palette="cool")
bp.set_ylabel("  Total Amount",fontsize=15)
plt.grid(axis="y",alpha=0.5)
## 4rd Plot
plt.subplot(2,2,4)
x=df.groupby("Gender")["Amount"].sum()
plt.pie(x=x,labels=["Female","Male"],autopct="%1.1f%%",radius=1.2,colors=["g","r"])
x=df.groupby("Gender")["Amount"].sum()
plt.pie([1],colors="w",radius=0.5)

plt.show()
```

#### Findings

##### BY ORDERS 
Most Customers Are Female 69.9 % (7832)
##### BY AMOUNT 
Most Customers Are Female 70.0 % (7832)\
The  purchasing power of Females is much greater than males.

#### Q Most Customers are Married or Unmarried ( Male & Female ) ?  

```python
plt.figure(figsize=(10,5))
# 1st Plot
df1=df[df["Gender"]=="Male"]
plt.subplot(2,2,1)
bp=sns.countplot(y="Marital_Status",data=df1)
for bars in bp.containers:
    bp.bar_label(bars)
plt.grid(axis="y",alpha=0.5)
plt.xlabel("")
plt.ylabel("Male",fontsize=15)
# 2nd Plot
ags=df1.groupby(by="Marital_Status")["Amount"].count()
plt.subplot(2,2,2)
plt.pie(x=ags,labels=["Married","Unmarried"],autopct="%1.1f%%",radius=1.2,colors=["r","g"])
plt.pie([1],colors="w",radius=0.5)
# 3rd plot
df2=df[df["Gender"]=="Female"]
plt.subplot(2,2,3)
bp=sns.countplot(y="Marital_Status",data=df2)
for bars in bp.containers:
    bp.bar_label(bars)
plt.grid(axis="y",alpha=0.5)
plt.ylabel("Female",fontsize=15)
# 4th Plot
ags=df2.groupby(by="Marital_Status")["Amount"].count()
plt.subplot(2,2,4)
plt.pie(x=ags,labels=["Married","Unmarried"],autopct="%1.1f%%",radius=1.2,colors=["r","g"])
plt.pie([1],colors="w",radius=0.5)

plt.show()
```

#### Findings  

Males :- 57.1 % Males are married\
Females :- 58.4 % Females are married\
we can see Most buyers are married. 

#### Q Maximum Customers belong to which Age Group?  

```python
df["Age Group"].unique()
```

```python
plt.figure(figsize=(11,5))

# plot1
plt.subplot(1,2,1)
cp=sns.countplot(x="Age Group",data=df,order=["0-17","18-25","26-35","36-45","46-50","51-55","55+"])
cp.set(title="Age Group  VS  Total Customers Count")
plt.grid(axis="y",alpha=0.5)
# Plot2
plt.subplot(1,2,2)
ages=df.groupby(["Age Group"],as_index=False)["Amount"].sum()
bp=sns.barplot(x="Age Group",y="Amount",data=ages)
bp.set(title="Age Group  VS  Total Amount")
plt.grid(axis="y",alpha=0.5)

plt.show()
```

#### Findings  

Maximum Customers belong to  26-35 Age Group

#### Q Age Group 26 - 35 maximum customers are ? ( Gender & Marital Status  ) 

```python
plt.figure(figsize=(11,5))
plt.subplot(1,2,1)
df2=df[df["Age Group"]=="26-35"]
ags=df2.groupby(by="Gender")["Marital_Status"].count()
plt.pie(x=ags,labels=["Female","Male"],autopct="%1.1f%%",radius=1.2,colors=["g","r"],explode=[0.02,0])
plt.pie([1],colors="w",radius=0.5)
plt.title("(Male vs Female )Age Group 26-35")

plt.subplot(1,2,2)
df2=df[df["Age Group"]=="26-35"]
ags=df2.groupby(by="Marital_Status")["Marital_Status"].count()
plt.pie(x=ags,labels=["Married","Umarried"],autopct="%1.1f%%",radius=1.2,colors=["r","g"],explode=[0,0.02],startangle=245)
plt.pie([1],colors="w",radius=0.5)
plt.title("(Married vs Unmarried) Age Group 26-35")

plt.show()
```

#### Findings

In Age Group 26 - 35\
Gender :- 72.0 % are Female\
Marital_Status :- 57.5 % are married


#### Q Maximum Customers belong to which State ? ( Total Orders & Total Amount ) 

```python
# PLOT 1
plt.figure(figsize=(11,4))
gbstate=df.groupby("State",as_index=False)["Orders"].sum().sort_values(by="Orders")
sns.barplot(x="State",y="Orders",data=gbstate)
plt.title("Total Orders Count",fontsize=15)
plt.xticks(rotation=70)
plt.grid(axis="y",alpha=0.4)
plt.show()
```

```python
# Plot 2
plt.figure(figsize=(11,4))
gbstate=df.groupby("State",as_index=False)["Amount"].sum().sort_values(by="Amount")
sns.barplot(x="State",y="Amount",data=gbstate)
plt.title("Total Sales Amount",fontsize=15)
plt.xticks(rotation=70)
plt.grid(axis="y",alpha=0.4)
plt.show()
```

#### Findings 

##### According to Orders :-
Top 3 States are Uttar Pradesh , Maharashtra , Karnataka

##### According to Amount :-
Top 3 States are Uttar Pradesh , Maharashtra , Karnataka


#### Q Maximum Customers belong to which Zone?  

```python
gbzone=df.groupby("Zone",as_index=False)["Amount"].sum().sort_values("Amount")
plt.figure(figsize=(10,4))
sns.barplot(y="Zone",x="Amount",data=gbzone)
plt.title("Zone  VS  Total Sales Amount ")
plt.grid(axis="x",alpha=0.3)
plt.show()
```

#### Findings 

Maximum Customers belong to Central & Sourthen Zone

#### Q Maximum Customers belong to which Age Group?  

```python
plt.figure(figsize=(10,5))
dfx=df.groupby(["Occupation"],as_index=False)["Orders"].count().sort_values("Orders")
sns.barplot(x="Occupation",y="Orders",data=dfx)
plt.xticks(rotation=45)
plt.show()
```

#### Q Which Product_Category is having maximum  Orders?  

```python
 plt.figure(figsize=(10,4))
dfx=df.groupby(["Product_Category"],as_index=False)["Orders"].count().sort_values("Orders")
sns.barplot(x="Product_Category",y="Orders",data=dfx)
plt.xticks(rotation=70)
plt.show()
```

#### Findings  

Product_Category Clothing & apparel ,Food ,Electronics & Gadagets are having maximum Orders

#### Q Which Product_Category is have generated maximum revenue ?  

```python
plt.figure(figsize=(10,4))
dfx=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values("Amount")
sns.barplot(x="Product_Category",y="Amount",data=dfx)
plt.xticks(rotation=70)
plt.show()
```

#### Findings

Product_Category which have generated maximum revenue are Food, Clothing & apparel, Electronics & Gadagets
